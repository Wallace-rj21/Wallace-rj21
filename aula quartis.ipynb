```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pesquisa_satisfacao.csv', sep=';')
df.head(3)
```

```python
media_satisfacao = df.groupby('Satisfação')['Tempo de Espera (min)'].mean().reset_index()
media_satisfacao['Tempo de Espera (min)'] = round(media_satisfacao['Tempo de Espera (min)'],2)
media_satisfacao
```

```python
df_anp = pd.read_csv('ANP2025.csv', sep=';')
df_anp
```

```python
df_anp_rj = df_anp.loc[df_anp['Estado - Sigla'] == 'RJ']
df_anp_rj.head(5)
```

```python
df_anp_rj['Valor de Venda'] = df_anp_rj['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
media_munic_rj = df_anp_rj.groupby('Municipio')['Valor de Venda'].mean().reset_index()
media_munic_rj = media_munic_rj.sort_values(by='Valor de Venda', ascending=False)
media_munic_rj
```

```python
array_valor_venda = np.array(media_munic_rj['Valor de Venda'])
```

```python
media = np.mean(array_valor_venda)
mediana = np.median(array_valor_venda)
q1 = np.quantile(array_valor_venda, 0.25)
q2 = np.quantile(array_valor_venda, 0.50)
q3 = np.quantile(array_valor_venda, 0.75)

print(f'A média é {media}')
print(f'A mediana é {mediana}')
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2, Mediana): {q2}')
print(f'Terceiro quartil (Q3): {q3}')

```

```python
munic_acima_q3 = media_munic_rj.loc[media_munic_rj['Valor de Venda'] > q3]
munic_acima_q3
```

```python
df_rj_etanol = df_anp.loc[(df_anp['Municipio'] == 'RIO DE JANEIRO') & (df_anp['Produto'] == 'ETANOL')]
df_rj_etanol.columns
```

```python
df_rj_etanol['Valor de Venda'] = df_rj_etanol['Valor de Venda'].str.replace(',','.').astype(float)
```
